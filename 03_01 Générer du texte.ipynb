```python
# Étape 2 : Importer les bibliothèques nécessaires
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from huggingface_hub import login
import os
```

# Les modèles

- Mistral-7B-Instruct : Prompt simple, explication, Q&R
- Mistral-7B : Fine-tuning ou R&D
- Mixtral-8x7B-Instruct : Très haute qualité de génération

# EN Python

```python
# Charger le tokenizer et le modèle
model_name = "mistralai/Mistral-7B-Instruct-v0.2"
login(token="XXXXXXXXXXXXXXXXXXXXXXx")

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype="auto")

# Définir le pipeline de génération
text_generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Prompt de départ
prompt = "Explique le concept de l'intelligence artificielle en termes simples"

# Génération du texte
output = text_generator(
    prompt,
    max_new_tokens=150,
    temperature=0.7,
    top_p=0.9,
    top_k=50,
    repetition_penalty=1.1,
    do_sample=True,
    eos_token_id=tokenizer.eos_token_id
)

# Affichage du résultat
print(output[0]['generated_text'])
```

```python
# Charger le tokenizer et le modèle
from huggingface_hub import InferenceClient

client = InferenceClient(model="mistralai/Mixtral-8x7B-Instruct-v0.1", token="XXXXXXXXXXXXXXXXXXXXXXx")

prompt = "Explique simplement ce qu'est l'apprentissage automatique."
response = client.text_generation(prompt, max_new_tokens=200)
print(response)
```
