# Q&A

Evaluer la performance : vérifier si le modèle donne la "bonne" réponse.

## Exact Match

Vérifie si la réponse du modèle correspond exactement à la référence attendue  

**evaluate.load("exact_match")**

## F1-Score

Mesure le chevauchement entre les mots prédits et attendus, utile si les réponses ne sont pas strictement identiques  

**evaluate.load("f1")**   
**sklearn.metrics.f1_score()**

# Résumé automatique

Générer un condensé fidèle et cohérent d’un texte long.

## ROUGE-1 / ROUGE-2 / ROUGE-L

Compare les n-grammes (les n mots consécutifs) du résumé généré à une ou plusieurs références.  
**evaluate.load("rouge")**

# Traduction automatique

Évaluer si la traduction générée est proche de la version humaine.


## BLEU Score

Mesure la similarité entre la traduction du modèle et une référence humaine, en comparant des blocs de mots (n-grammes).  

**evaluate.load("bleu")**  
**sacrebleu.corpus_bleu()** pour usage avancé

# Classification

Prédire une catégorie correcte à partir d’un texte.



## Accuracy

Pourcentage de bonnes prédictions sur l’ensemble du dataset.

**sklearn.metrics.accuracy_score()**

## Precision / Recall / F1 Score

Équilibre entre taux de vrais positifs et faux positifs ; utile en cas de classes déséquilibrées.

**sklearn.metrics.classification_report()**  
**f1_score()**  
**precision_score()**  

## Confusion Matrix

Affiche les erreurs par classe, utile pour identifier les biais du modèle.

**sklearn.metrics.confusion_matrix()**

# Génération

Générer un texte pertinent, fluide, dans le bon ton, sans vérité absolue.


## Evaluation humaine (indispensable)

On évalue la fluidité, la pertinence, le style, subjectivement ou avec une grille.

## BERTScore

Compare le sens (embedding sémantique) entre la sortie du modèle et une référence.  
**bert_score.score()**  

## MAUVE

Compare la distribution du texte généré à celle de textes humains.  
**mauve.compute_mauve()**
